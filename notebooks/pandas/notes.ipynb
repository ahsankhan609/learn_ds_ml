### creating series

Series is a **one-dimensional** array-like object containing an array of **various data-types**(integer, float, string, boolean, etc.) and an **associated array of data labels**, called its **index**.

Series is a buit-in Class in pandas.

```python
# from list

countries:list[str] = ['India', 'USA', 'Russia', 'China', 'Brazil']
pd.Series(countries)
```

series have 2 things. One is data and other is index.

```python
# from integers using np array

runs:np.ndarray = np.array([1,2,3,4,5])
pd.Series(runs, name="Serieswith_numbers")
```

```python
# creating series with index
my_series = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
my_series
```

```python
subjects:list[str] = ['Math', 'Science', 'English', 'Social', 'Coding']
marks:np.ndarray = np.array([95, 56, 78, 90, 96])

marks_series = pd.Series(marks, index=subjects)
marks_series
```

```python
# setting a name for the series

marks_series = pd.Series(marks, index=subjects, name='StudentMarks')
marks_series
```

```python
# creating series from dictionary

marks_dict: dict[str, int] = {'Math': 95, 'Science': 56, 'English': 78, 'Social': 90, 'Coding': 96}
marks_series:Series = pd.Series(marks_dict, name='StudentMarks_with_Dictionary')
marks_series
```

### series attributes

```python
# series attributes apply to a Series, not a DataFrame
marks_series = pd.Series(marks_dict, name='StudentMarks_with_Dictionary')
```

```python
# marks_series.size

marks_series.size # returns the number of elements in the series
```

```python
# marks_series.dtype

marks_series.dtype # returns the data type of the series
```

```python
# marks_series.nbytes

marks_series.nbytes # returns the number of bytes in the series
```

```python
# marks_series.name

marks_series.name # returns the name of the series
```

```python
# marks_series.is_unique

marks_series.is_unique # returns True if the all items in the series are unique, otherwise False
```

```python
marks_series.index
```

```python
marks_series.values # returns the values of the series in an numpy array
```
